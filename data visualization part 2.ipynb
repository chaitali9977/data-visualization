```python
import seaborn as snr
```

```python
new_data=snr.load_dataset('tips')
```

```python
new_data.head()
```

```python
#relplot
snr.relplot(x='total_bill',y='tip',data=new_data,color='purple')
```

```python
snr.relplot(x='total_bill',y='tip',data=new_data,hue='smoker',style='time')
```

```python
snr.relplot(x='total_bill',y='tip',data=new_data,hue='smoker',style='time',size='size')
```

```python
snr.relplot(x='total_bill',y='tip',data=new_data,hue='smoker',style='sex',size='size',col='time')
```

```python
snr.relplot(x='total_bill',y='tip',data=new_data,hue='smoker',style='time',size='size',col='sex')
```

```python
#catplot
snr.catplot(y='day',x='total_bill',data=new_data)
```

```python
#box plot# we r not using here style because it will be either in cross shape or circle shape but this is already in box shape 
snr.catplot(y='tip',x='day',kind='box',data=new_data,hue='sex')
```

```python
#violin plot
snr.catplot(x='total_bill',y='day',hue='sex',kind='violin',data=new_data)
```
